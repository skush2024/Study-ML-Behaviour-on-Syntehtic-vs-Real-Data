```python
import pandas as pd 
```

```python
path = "/Users/skush/CodeX/Stress-Level-Classifier/data/processed/cleaned_survey_responses.csv"
```

```python
df = pd.read_csv(path)
df = df.loc[:, ~df.columns.str.contains('^Unnamed')]
```

```python
df.head()
```

```python
irrelevant_columns = ['Device_Type']
```

```python
df = df.drop(columns=irrelevant_columns)
```

```python
df["Gender"] = df["Gender"].str.lower()
df["Occupation"] = df["Occupation"].str.lower()
```

```python
df["Stressed"] = df["Stress_Level"].apply(lambda x: 1 if x > 5 else 0)
```

```python
df.head()
```

```python
df["Occupation"] = df["Occupation"].str.split(";")
```

```python
df = df.explode("Occupation")
```

```python
df.shape
```

```python
df = pd.get_dummies(df)
```

```python
df.head()
```

```python
df.columns
```

```python
df = df.rename(columns={
    'Daily_Phone_Hours' : 'Daily_Screen_Hours', 
    'Caffeine_Intake_Cups' : 'Caffeine_Cups_Per_Day'
})
```

```python
df.to_csv("/Users/skush/CodeX/Stress-Level-Classifier/data/processed/survey_data_labelled.csv")
```
